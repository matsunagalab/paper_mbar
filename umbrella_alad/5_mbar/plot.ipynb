```julia
using Plots, DelimitedFiles
```

```julia
d_wham = readdlm("wham.pmf")
```

```julia
d_mbar = readdlm("mbar.pmf")
```

```julia
plot(d_mbar[:, 1], d_mbar[:, 2] .- d_mbar[1, 2], linewidth=3, label="MBAR", color=2)
p = plot!(d_wham[:, 1], d_wham[:, 2] .- d_wham[1, 2], linewidth=3, color=1, linestyle=:dash, label="WHAM",
          framestyle=:box, 
          xlabel="angle [deg]", ylabel="potential of mean force [kcal/mol]",
          xtickfontsize=12, ytickfontsize=12, legendfontsize=12,
          xlabelfontsize=18, ylabelfontsize=18, dpi=200, size=(700, 550))

#savefig("plot.png")
```

```julia
savefig(p, "plot.png")
```
