```julia
using Plots, DelimitedFiles
```

```julia
d_single = readdlm("single.pmf")
```

```julia
d_mbar = readdlm("dist.pmf")
```

```julia
p = plot(d_mbar[:, 1], d_mbar[:, 2], linewidth=3, color=:2, dpi=150, box_style=:box, label="MBAR", size=(500, 400), legend=:bottomleft)
plot!(d_single[:, 1], d_single[:, 2], linewidth=3, color=:1, linestyle=:dash, label="Single temperature", 
      framestyle=:box, 
      xlabel="distance [Å]", ylabel="potential of mean force [kcal/mol]",
      xtickfontsize=12, ytickfontsize=12, legendfontsize=12,
      xlabelfontsize=18, ylabelfontsize=18, dpi=200, size=(700, 550))
xlims!(1.5, 13); ylims!(-0.2, 4.2)
xaxis!("distance [Å]"); yaxis!("potential of mean force [kcal/mol]")
```

```julia
savefig(p, "plot.png")
```
