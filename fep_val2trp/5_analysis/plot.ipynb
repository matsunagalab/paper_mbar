```julia
plot(lambda, d_mean, ribbon=2.0 .* d_std, fillalpha=0.3, color=2,
      linewidth=3, label="MBAR", legend=:topleft, 
      xtickfontsize=12, ytickfontsize=12, legendfontsize=12,
      framestyle=:box, 
      xlabel="lambda", ylabel="free energy difference [kcal/mol]",
      xlabelfontsize=18, ylabelfontsize=18, dpi=200)

p = plot!(lambda, exp_mean, color=1,
      linewidth=3, linestyle=:dash, label="EXP", size=(700, 550))
```
